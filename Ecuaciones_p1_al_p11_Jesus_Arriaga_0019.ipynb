#### Ecuacion 1
$x + 8 = 15$
## Paso 1: Identificamos que el $+8$ debe moverse al lado derecho para dejar sola a la $x$.
## Al cruzar el signo $=$, cambia a resta.
## Paso 2: Escribimos la nueva igualdad: $x = 15 - 8$.
## Paso 3: Resolvemos la operación final: $x = 7$.

*   Elemento de la lista
*   Elemento de la lista



```python
# 1. Ecuación: x + b = c
print("--- Resolviendo: x + b = c ---")
b = float(input("Ingresa el número que suma a x (ej. 8): "))
c = float(input("Ingresa el resultado total (ej. 15): "))

# Proceso: x = c - b
resultado = c - b

print(f"\nPasos:")
print(f"x = {c} - {b}")
print(f"RESULTADO: x = {resultado}")
```

#### Ecuacion 3
$3x + 4x = 21$
### Paso 1: Agrupamos los términos semejantes en el lado izquierdo ($3x$ y $4x$).
### Sumamos sus coeficientes: $3 + 4 = 7$.
### Paso 2: La ecuación simplificada es $7x = 21$.
## Paso 3: El $7$ que multiplica a la $x$ pasa dividiendo al $21$.
## Paso 4: Resolvemos: $x = 3$.

```python
# 2. Ecuación: Ax + Bx = C
print("--- Resolviendo: Ax + Bx = C ---")
a = float(input("Ingresa el primer coeficiente de x (ej. 3): "))
b = float(input("Ingresa el segundo coeficiente de x (ej. 4): "))
c = float(input("Ingresa el resultado total (ej. 21): "))

# Proceso: (A + B)x = C
suma_coef = a + b
resultado = c / suma_coef

print(f"\nPasos:")
print(f"Agrupando x: ({a} + {b})x = {c}")
print(f"{suma_coef}x = {c}")
print(f"x = {c} / {suma_coef}")
print(f"RESULTADO: x = {resultado}")
```

#### Ecuacion 5
$6a - 2a + a = 20$
### Paso 1: Agrupamos los términos con la letra $a$.
#### Recuerda que $a$ es lo mismo que $1a$.
### Paso 2: Operamos: $6a - 2a = 4a$, y luego $4a + 1a = 5a$.
### Paso 3: La ecuación queda como $5a = 20$.
### Paso 4: El $5$ pasa dividiendo: $a = 20 / 5$. Resultado: $a = 4$.

```python
# Ecuación: Aa - Ba + Ca = D
print("--- Resolviendo Ecuación 3 (6a - 2a + a = 20) ---")

c1 = float(input("Primer coeficiente de 'a' (ej. 6): "))
c2 = float(input("Segundo coeficiente (el que resta, ej. 2): "))
c3 = float(input("Tercer coeficiente (ej. 1): "))
total = float(input("Resultado final (ej. 20): "))

# Agrupación: (c1 - c2 + c3)a = total
coef_final = c1 - c2 + c3
a_resultado = total / coef_final

print(f"\nAgrupando términos semejantes:")
print(f"({c1} - {c2} + {c3})a = {total}")
print(f"{coef_final}a = {total}")
print(f"Resultado: a = {a_resultado}")
```

#### Ecuacion 7
$4x + 3 = 2x + 11$
## Paso 1: Agrupamos las $x$ en la izquierda.
## El $2x$ pasa como $-2x$.Paso 2:
## Agrupamos los números en la derecha.
## El $+3$ pasa como $-3$.
### Paso 3: La ecuación se ve así: $4x - 2x = 11 - 3$.
## Paso 4: Resolvemos las restas en ambos lados: $2x = 8$.
## Paso 5: El $2$ pasa dividiendo: $x = 8 / 2$. Resultado: $x = 4$

```python
# Ecuación: Ax + B = Cx + D
print("--- Resolviendo Ecuación 4 (4x + 3 = 2x + 11) ---")

A = float(input("Coeficiente x izquierdo (ej. 4): "))
B = float(input("Número solo izquierdo (ej. 3): "))
C = float(input("Coeficiente x derecho (ej. 2): "))
D = float(input("Número solo derecho (ej. 11): "))

# Agrupación: (A - C)x = D - B
lado_izquierdo = A - C
lado_derecho = D - B
x = lado_derecho / lado_izquierdo

print(f"\nPasos de agrupación:")
print(f"{A}x - {C}x = {D} - {B}")
print(f"{lado_izquierdo}x = {lado_derecho}")
print(f"Resultado: x = {x}")
```

#### Ecuacion 9
$9m + 4 = 5m + 20$
## Paso 1: Movemos términos semejantes:
##  El $5m$ a la izquierda y el $4$ a la derecha.
## Paso 2: Escribimos la agrupación: $9m - 5m = 20 - 4$.
## Paso 3: Simplificamos ambos lados: $4m = 16$.
### Paso 4: Despejamos la $m$ dividiendo: $m = 16 / 4$. Resultado: $m = 4$

```python
# Ecuación: Am + B = Cm + D
print("--- Resolviendo Ecuación 5 (9m + 4 = 5m + 20) ---")

a_m = float(input("Coeficiente m izquierdo (ej. 9): "))
b_num = float(input("Número izquierdo (ej. 4): "))
c_m = float(input("Coeficiente m derecho (ej. 5): "))
d_num = float(input("Número derecho (ej. 20): "))

# Agrupación: (a_m - c_m)m = d_num - b_num
resta_m = a_m - c_m
resta_num = d_num - b_num
m = resta_num / resta_m

print(f"\nPasos de agrupación:")
print(f"{a_m}m - {c_m}m = {d_num} - {b_num}")
print(f"{resta_m}m = {resta_num}")
print(f"Resultado: m = {m}")
```

#### Ecuacion 11
$6a + 5 = 2a + 21$
## Paso 1: Llevamos las letras $a$ a un lado: $6a - 2a$.
### Paso 2: Llevamos los números al otro: $21 - 5$.
## Paso 3: La ecuación queda: $4a = 16$.
### Paso 4: Resolvemos la división: $a = 16 / 4$.
## Resultado: $a = 4$.

```python
# Ecuación: Aa + B = Ca + D
print("--- Resolviendo Ecuación 6 (6a + 5 = 2a + 21) ---")

coef1 = float(input("Coeficiente 'a' izquierdo (ej. 6): "))
num1 = float(input("Número izquierdo (ej. 5): "))
coef2 = float(input("Coeficiente 'a' derecho (ej. 2): "))
num2 = float(input("Número derecho (ej. 21): "))

# Agrupación: (coef1 - coef2)a = num2 - num1
final_a = coef1 - coef2
final_num = num2 - num1
resultado_a = final_num / final_a

print(f"\nPasos de agrupación:")
print(f"{coef1}a - {coef2}a = {num2} - {num1}")
print(f"{final_a}a = {final_num}")
print(f"Resultado: a = {resultado_a}")
```
